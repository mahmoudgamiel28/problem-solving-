The center of the rangoli has the first alphabet letter *a*, and the boundary has the $N$<sup>th</sup>  alphabet letter (in alphabetical order).

**Function Description**

Complete the *rangoli* function in the editor below.

*rangoli* has the following parameters:

- *int size:* the size of the rangoli

**Returns**

- *string:* a single string made up of each of the lines of the rangoli separated by a newline character (\n)

**Input Format**

Only one line of input containing  $size$, the size of the rangoli.

**Constraints**

$0< size <27$

**Sample Input**

```
5
```

**Sample Output**

```
--------e--------
------e-d-e------
----e-d-c-d-e----
--e-d-c-b-c-d-e--
e-d-c-b-a-b-c-d-e
--e-d-c-b-c-d-e--
----e-d-c-d-e----
------e-d-e------
--------e--------
```


````python
def print_rangoli(n):
    from string import ascii_lowercase as chars
    heap = [(('-'.join(chars[i:n])[::-1]+'-'.join(chars[i:n])[1:])).center(4*n-3, '-') for i in range(n)]
    print(*(heap[::-1]+heap[1:]), sep="\n")
````

````python
n = int(input())
print_rangoli(n)
````
